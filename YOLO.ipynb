```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

# utilities

```python
def prediction(model_type,img_path,display_result=False,task=None):

    model = YOLO(model_type)
    result = model(img_path, save=True, conf=0.7)

    for r in result:
        pred_img_path = f"{r.save_dir}/{img_path}"
        pred = cv2.cvtColor(cv2.imread(pred_img_path), cv2.COLOR_BGR2RGB)
        plt.imshow(pred)
        plt.show()
        plt.axis(False)
        plt.title(f"YOLOl11 - {task}")

    if display_result:
        print(result)
```

# Object Detection

```python
model_type = "yolo11l.pt"
```

## inference


```python
img_path = "football.jpg"
prediction(model_type, img_path, task = "Object Detection")
```

```python
img_path = "zebra.jpg"
prediction(model_type, img_path, display_result= True, task = "Object Detection")
```

```python
img_path = "monitor.jpg"
prediction(model_type, img_path, task = "Object Detection")
```
